```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 通过np.linspace方式生成x，它是包含50个元素的数组，均匀分布在区间[0,2pi]上
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
plt.plot(x,y)
# m代表颜色品红，--代表虚线，更多见文档
plt.plot(x,2 * y,'m--')
plt.show()
```

```python
# 自己创建一个figure，大小为(6, 3)
plt.figure(figsize=(6, 3))
plt.plot(x, y)
plt.plot(x, y * 2)
# 设置标题
plt.title('sin(x) and 2sin(x)')
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, y * 2)
# 设置坐标轴的范围
plt.xlim(0, np.pi * 2)
plt.ylim(-3, 3)
# 设置坐标轴的名称
plt.xlabel('X')
plt.ylabel('Y')
# 设置坐标轴的刻度
plt.xticks((0, 0.5*np.pi, 1*np.pi, 1.5*np.pi, 2*np.pi))
plt.yticks((-2,-1.5,-1,-0.5,0,0.5,1,1.5,2))
plt.show()
```

```python
# 设置label和legend
plt.plot(x, y, label='sin(x)')
plt.plot(x, y * 2, label='2sin(x)')
# 显示图例
plt.legend(loc='best')
plt.show()
```

```python
plt.plot(x, y)
x0 = np.pi
y0 = 0
# 画出标注点
plt.scatter(x0, y0, s=50)
# 用plt.annotate对点进行标注，xy是标注点的坐标，xycoords='data'代表基于点的坐标值来选择位置
# xytext(+30,-30)代表标注位置的text与xy坐标的偏差，此处指沿x轴正向移动30，y轴负向移动30,
# arrowprops是对图中箭头类型和箭头弧度的设置，需要用dict形式传入
plt.annotate('sin(pi)=%s' %y0, xy=(np.pi, 0), xycoords='data', xytext=(+30, -30),textcoords='offset points',fontsize=16,\
            arrowprops=dict(arrowstyle='->'))
# 用plt.text添加注释，(0.5,-0.25)是标注位置的坐标，字体颜色为r（红色）
plt.text(0.5, -0.25, 'sin(pi) = 0', fontdict={'size':16,'color':'r'})
plt.show()
```

```python
# 使用子图
# subplot第一个参数代表子图的总行数，第二个代表总列数，第三个代表当前子图的活跃区域
ax1 = plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), 'r')

ax2 = plt.subplot(2, 2, 2, sharey=ax1)
plt.plot(x, 2 * np.sin(x), 'g')

ax3 = plt.subplot(2, 2, 3)
plt.plot(x, np.cos(x), 'b')

ax4 = plt.subplot(2, 2, 4)
plt.plot(x, 2 * np.cos(x), 'y')

plt.show()
```

```python
# 第一行一张子图，第二行三张子图
ax1 = plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), 'r')

ax2 = plt.subplot(2, 3, 4)
plt.plot(x, 2 * np.sin(x), 'g')

ax3 = plt.subplot(2, 3, 5, sharey=ax2)
plt.plot(x, np.cos(x), 'b')

ax4 = plt.subplot(2, 3, 6, sharey=ax2)
plt.plot(x, 2 * np.cos(x), 'y')

plt.show()
```

```python
k = 500
# 随机生成含500个数的列表x,y
x = np.random.rand(k)
y = np.random.rand(k)
# 随机生成每个点的大小
size = np.random.rand(k) * 50
# 生成每个点的颜色大小
color = np.arctan2(x, y)
# 绘制散点图
plt.scatter(x, y, s=size, c=color)
#添加颜色栏
plt.colorbar()

plt.show()
```

```python
k = 10
x = np.arange(k)
y = np.random.rand(k)
#绘制柱状图
plt.bar(x, y)
for x, y in zip(x, y):
    #标注数值
    # ha='center'横向居中对齐， va='bottom'纵向底部对齐
    plt.text(x, y, '%.2f' %y, ha='center', va='bottom')

plt.show()
```
